```python
import numpy as np 
import pandas as pd
```

```python
customers={"Customerıd":[1,2,3,4],
           "Customer Name":["yalçındağ","mustafa","ede turzim","kiya inşaat"],
           "Customer City":["istanbul","istanbul","ankara","ankara"],
           "age":[np.nan,23,np.nan,56]}
```

```python
customerDF= pd.DataFrame(customers,columns=["Customerıd","Customer Name","Customer City","age"])
customerDF
```

```python
sales={"Saleıd":[23,52,41,36],
       "Customerıd":[1,2,3,4],
       "Product":["çimento","bilet","bilet","çimento"],
       "Amounnt":[541,750,123,np.nan]}
saleDF=pd.DataFrame(sales)
saleDF
```

```python
salesMart = pd.DataFrame({
    "SaleID": [1,5,7,8],
    "CustomerID": [99,104,206,207],
    "Product": ["Laptop","Tablet","Telefon","Saç maşası"],
    "Amount": [45000, 12860, 37000, 3100],
    "Month": ["2026-03", "2026-03", "2026-03", "2026-03"]
})

salesEyl = pd.DataFrame({
    "SaleID": [2,6,7],
    "CustomerID": [100,105,207],
    "Product": ["Kulaklık","Laptop","Saç maşası"],
    "Amount": [2300,72950, 3100],
    "Month": ["2026-09", "2026-09", "2026-09"]
    
})
```

```python
aybirleştirme=pd.concat([salesMart,salesEyl], ignore_index=True) 
aybirleştirme                                 # yukarıdaki indexi sıfırlar
```

```python
aybirleştirme["KDV"] = aybirleştirme["Amount"] * 1.20
print("KDV eklenmiş satış tutarı: ")
aybirleştirme

#apply() kullanımı -> Satış tutarına göre kategori oluşturalım
def satisKategori(amount):
    if amount >= 40000:
        return "Premium"
    elif amount >= 25000:
        return "Gold"
    elif amount >= 10000:
        return "Standard"
    else:
        return "Basic"
aybirleştirme["Category"] = aybirleştirme["Amount"].apply(satisKategori)
print("Kateogri eklenmiş hali: ")
aybirleştirme
```

```python
# Temizlenmiş ve analiz edilmiş tabloyu excel'e yaz
aybirleştirme.to_excel("final_sales_report.xlsx", index=False) #dosyayı oluşturdul
print("Excel dosyası yazıldı.")

#Excel dosyasını okuyalım
salesFromExcel = pd.read_excel("final_sales_report.xlsx") #dosyayı tekrar yükledik 
print("Excel'den okunan veri")
display(salesFromExcel)
```
